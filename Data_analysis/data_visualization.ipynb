# Analisi dataset TRAINING

```python
# Caricamento dataset training

train_path_X = "Dataset\Train\X_train.csv"
train_path_Y = "Dataset\Train\Y_train.csv"

df_x = pd.read_csv(train_path_X)
df_y = pd.read_csv(train_path_Y)
```

```python
print(f"Lunghezza df_y: {len(df_y)}")
df_y.head()
```

```python
df = pd.concat([df_x, df_y], axis=1)
df = df.drop(columns=[col for col in df.columns if col == 'id'])
df['np/ng'] = df['np'] / df['ng']
print(f"Lunghezza df: {len(df)}")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definisci una lista con i nomi delle variabili e dei titoli
features = ['trq_measured', 'oat', 'mgt', 'pa', 'ias', 'np', 'ng', 'trq_margin', 'np/ng']
titles = ['Torque Measured', 'Outside Air Temperature', 'Measured Gas Temperature', 'Pressure Altitude', 
          'Indicated Airspeed', 'Power Turbine Speed', 'Gas Generator Speed', 'Torque Margin', 'np/ng']
y_labels = titles

# Crea una figura con 3 righe e 3 colonne (ultima cella vuota)
fig, axes = plt.subplots(3, 3, figsize=(18, 18))

# Itera attraverso le variabili e i subplot
for i, feature in enumerate(features):
    # Calcola la posizione corretta nel subplot (riga, colonna)
    row = i // 3
    col = i % 3
    
    # Aggiungi la colonna 'source' per identificare la provenienza dei dati
    df_fault['source'] = 'Faulty'
    df_no_fault['source'] = 'Fault-free'

    # Seleziona la variabile e la colonna 'source' per entrambi i dataframe
    df_fault_selected = df_fault[[feature, 'source']]
    df_no_fault_selected = df_no_fault[[feature, 'source']]

    # Unisci i due dataframe in uno solo
    df_combined = pd.concat([df_fault_selected, df_no_fault_selected])

    # Crea il boxplot per la variabile corrente
    sns.boxplot(x='source', y=feature, data=df_combined, ax=axes[row, col])

    # Imposta il titolo e le etichette per ogni subplot
    axes[row, col].set_title(f'Box-plot di {titles[i]}', fontsize=14)
    axes[row, col].set_xlabel('', fontsize=12)
    axes[row, col].set_ylabel(y_labels[i], fontsize=12)



# Regola il layout per evitare sovrapposizioni
plt.tight_layout()

# Mostra il grafico
plt.show()
```

Dalle matrice di correlazione, si pu notare come:

-   **MATRICE DI CORRELAZIONE FAULTY + FAULT-FREE:** 
    
    -   **La correlazione del 64% tra mgt (Measured Gas Temperature) e trq_measured (Torque Measured)** 
    indica una relazione positiva moderata tra la temperatura dei gas di scarico e la coppia misurata. In generale, un aumento della coppia richiede un incremento della produzione di energia, il che comporta un aumento della combustione e, di conseguenza, una maggiore temperatura nei gas di scarico. Tuttavia, la correlazione non  pari a 1, il che significa che l'aumento della coppia non porta a un aumento proporzionale della temperatura dei gas di scarico. Questo accade perch il sistema  progettato per mantenere la temperatura dei gas all'interno di un range sicuro. Infatti, se la temperatura dei gas di scarico continuasse ad aumentare all'aumentare della coppia, ci potrebbe danneggiare i componenti del motore o compromettere le prestazioni del veicolo. Per questo motivo, il condotto del gas di scarico e il sistema di gestione del motore sono progettati in modo tale da limitare l'incremento della temperatura oltre una certa soglia, evitando danni al sistema. *Tale corellazione  presente anche nelle matrici di correlazione FAULTY e FAULT-FREE*.

    
    -   **ng(Gas generator speed) e np(Power turbine speed) correlate al -79%**
        La correlazione negativa tra *ng* e *np*  probabilmente dovuta alla dinamica di come le due turbinas lavorano insieme per mantenere un equilibrio tra potenza e carico, e un aumento della velocit di uno potrebbe comportare la necessit di ridurre la velocit dell'altro per evitare sovraccarichi o inefficienze.

-   **MATRICE DI CORRELAZIONE FAULTY:**

    -   **oat(Outside Air Temperature) e np(Power Turbine Speed) correlate al 63 %**
        In caso di guasto, il motore a reazione non riuscir pi a compensare variazioni di temperatura, ci significa che all'aumentare/diminuire della temperatura auenter/diminuir la coppia del motore del veivolo.

    -   **ng(Gas Generator Speed) e oat(Outside Air Temperature) correlate al -66%**
        Con temperature pi alte, la velocit del generatore di gas diminuisce.
        Spiegando con un esempio, all'aumentare della temperatura, aumenta la potenza della turbina, si dovrebbe dunque produrre piu energia, ma siccome  presente un guasto, allora l'energia prodotta  minore e di conseguenza anche la velocit dei gas di scarico diminuisce. 

    -   **ias(Indicated Airspeed) e oat(Outside Air temperature) correlati al -42%**
        Questa correlazione  spiegabile intersecando i ragionamenti precedenti. In caso di guasto, un aumento della temperatura, comporta una riduzione di energia ed ng, quindi, la velocit del veivolo diminuisce.

    -   **np/ng e oat correlati al 68%**
        Se *np*  correlato con *oat* ed *ng*  correlato con *oat*, allora anche il loro rapporto sar correlato.

-   **MATRICE DI CORRELAZIONE FAULT-FREE:**

    E' interessante notare, come *ias*, *np* ed *ng* Non siano correlate con *oat*, a differenza di quanto accade nella matrice di correlazione FAULTY. Il comportamento anomalo, mette in correlazione variabili che prima non lo erano(o lo erano poco).


**Nella selezione delle feature, un buon punto di partenza, sarebbe quello di valutare ed addestrare i modelli per le feature oat, np/ng, ias. Infatti tali variabili sono "particolari", in quanto alcune correlazioni che sono fortemente presenti nei casi FAULTY non sono presenti o sono presenti in negativo, nei casi FAULT-FREE e viceversa**

**Idea**: Andando a clusterizzare i datafram con guasti, si potrebbe anche individuare le classi di guasto
